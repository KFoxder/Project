#### Part 3 )  Prepare the raw json ad data for prediction, transforming the ad / nonad label to a 1 / 0 numeric value, and change the string values in the json array to floats.

```python
import json as json

file = open("adFiltered.json","r")
file2 = open("adFilteredTrans.csv","w+")

dataTransformed = []
for line in file:

    data = json.loads(line)
    if data[-1] == 'ad.':
        data[-1] = int(1)
    elif data[-1] == 'nonad.':
        data[-1] = int(0)
        
    for i,v in enumerate(data[0:-1]):
        data[i] = float(data[i])
    dataTransformed.append(data)
    file2.write(str(json.dumps(data)).replace('[','').replace(']','')+"\n")


file.close()
file2.close()
!wc -l adFilteredTrans.csv
!head -n1 adFilteredTrans.csv
```

#### Part 4) Train a logistic regression model and a decision tree classifier model on the data prepared in step 3. What is the accuracy of these models when applied back on the training data?

Logistic Regression Model: 

```python
from sklearn import linear_model
from numpy import genfromtxt
dataset = genfromtxt(open('adFilteredTrans.csv','r'), delimiter=',', dtype='f8') 

target = [x[-1] for x in dataset]
train = [x[0:-1] for x in dataset]


logModel = linear_model.LogisticRegression()
logModel.fit(train,target)

predictions = logModel.predict(train)

wrongPred = 0

for i, v in enumerate(predictions):
    if predictions[i] != target[i]:
        wrongPred+=1

print("The logisticRegression Model got "+str(wrongPred)+" predictions wrong.")
print("Accuracy was :"+str(float(1-float(float(wrongPred))/float(len(predictions))))+"%")
```

Decision Tree Model:

```python
from sklearn import tree
from numpy import genfromtxt
dataset = genfromtxt(open('adFilteredTrans.csv','r'), delimiter=',', dtype='f8') 

target = [x[-1] for x in dataset]
train = [x[0:-1] for x in dataset]


DTModel = tree.DecisionTreeClassifier()
DTModel.fit(train,target)

predictions = DTModel.predict(train)

wrongPred = 0

for i, v in enumerate(predictions):
    if predictions[i] != target[i]:
        wrongPred+=1

print("The Decision Tree Model got "+str(wrongPred)+" predictions wrong.")
print("Accuracy was :"+str(float(1-float(float(wrongPred))/float(len(predictions)))))
```

#### Part 5 ) This question is focused on the more realistic setting where the model is applied to data that wasn’t available at training time. Segment the data into training and test splits. Hint: use cross_validation.train_test_split. Split the data into {5%, 10%, …, 95%} training, in each case, using the remainder as testing.<br/><br/> Part 6 )  Build logistic regression and decision tree classifiers on the training sets generated in question 5. What is the accuracy of these models when they are applied to the testing side of each split? Plot the performance of both models, with the % of data in the training set on the x-axis and the model’s accuracy on the y-axis.

```python
import numpy as np
from sklearn import cross_validation , metrics
from sklearn import tree
from numpy import genfromtxt

dataset = genfromtxt(open('adFilteredTrans.csv','r'), delimiter=',', dtype='f8') 

target = [x[-1] for x in dataset]
train = [x[0:-1] for x in dataset]

logModel = linear_model.LogisticRegression()
DTModel = tree.DecisionTreeClassifier()

logVal = []
treeVal = []
XVals = []

for percTest in np.linspace(0.05, 0.95, num=19):
    XVals.append(percTest)
    
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(train, target, test_size=percTest, random_state=0)

    DTModel.fit(X_train,y_train)
    logModel.fit(X_train,y_train)
    y_pred_log = logModel.predict(X_test)
    y_pred_tree = DTModel.predict(X_test)

    logVal.append(metrics.accuracy_score(y_test, y_pred_log))
    treeVal.append(metrics.accuracy_score(y_test, y_pred_tree))
    
    print "Test Percent "+str(percTest)
    print "Accuracy Tree : %f"% metrics.accuracy_score(y_test, y_pred_tree)
    print "Accuracy Log Reg : %f"% metrics.accuracy_score(y_test, y_pred_log)
    print ""
```

#### Part 6) Plot

```python
%matplotlib inline
from matplotlib import pyplot as plt

plt.title("Log Reg v. Tree Accuracy")
plt.plot(XVals,treeVal,color='green',label='Decision Tree')
plt.plot(XVals,logVal,color='red',label='Log. Reg.')
plt.legend(loc='lower left')
plt.xlabel("Perc of Data used for Testing")
plt.ylabel("Accuracy Percent")
```

#### Part 7 ) Extend step 6. using one other classification model of your choice

```python
import numpy as np
from sklearn import cross_validation , metrics
from sklearn import tree
from numpy import genfromtxt
from matplotlib import pyplot as plt
from sklearn.naive_bayes import GaussianNB

dataset = genfromtxt(open('adFilteredTrans.csv','r'), delimiter=',', dtype='f8') 

target = [x[-1] for x in dataset]
train = [x[0:-1] for x in dataset]

logModel = linear_model.LogisticRegression()
DTModel = tree.DecisionTreeClassifier()
GNBModel = GaussianNB()


gnbVal = []
logVal = []
treeVal = []
XVals = []

for percTest in np.linspace(0.05, 0.95, num=19):
    XVals.append(percTest)
    
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(train, target, test_size=percTest, random_state=0)

    GNBModel.fit(X_train,y_train)
    DTModel.fit(X_train,y_train)
    logModel.fit(X_train,y_train)
    y_pred_log = logModel.predict(X_test)
    y_pred_tree = DTModel.predict(X_test)
    y_pred_gnb = GNBModel.predict(X_test)

    logVal.append(metrics.accuracy_score(y_test, y_pred_log))
    treeVal.append(metrics.accuracy_score(y_test, y_pred_tree))
    gnbVal.append(metrics.accuracy_score(y_test,y_pred_gnb))
    



plt.title("Log Reg v. Tree v. Naive Bayes Accuracy")
plt.plot(XVals,treeVal,color='green',label='Decision Tree')
plt.plot(XVals,logVal,color='red',label='Log. Reg.')
plt.plot(XVals,gnbVal,color='blue',label='Naive Bayes')
plt.legend(loc='lower left')
plt.xlabel("Perc of Data used for Testing")
plt.ylabel("Accuracy Percent")
```
