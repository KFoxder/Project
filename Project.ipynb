```python
file = open("training.csv","r")

isGood = 0
isBad = 0

for line in file:
    columns = line.split(',')
    if columns[1] == '0':
        isGood +=1
    if columns[1] == '1':
        isBad+=1
file.close()
percent = float(float(isBad)/float(isGood+isBad))
print("Baseline percentage of bad pruchases of cars is :")
print(str(percent))
```

```python
## Removing lines will NULL values except for in Two Columns 


file = open("training.csv","r")
cleanFile = open("trainingClean.csv","w+")
rmLines=0
for line in file:
    columns = line.split(',')
    if "NULL" not in columns[0:25] and "NULL" not in columns[28:] :
        cleanFile.write(line)
    else:
        rmLines+=1  
print("Removed "+str(rmLines)+" NULL lines from file.")
cleanFile.close()
file.close()
!wc -l trainingClean.csv
!wc -l training.csv
```

```python
#Clean up the PRIMEUNIT & AUGCCART Column to numeric 
file = open("trainingClean.csv","r")
cleanFile = open("trainingClean2.csv","w+")
cleanFile.write(file.next())
for line in file:
    columns = line.split(',')
    cleanFile.write(line.replace("NULL",str(int(0))).replace("YES",str(int(2))).replace("NO",str(int(1))).replace("GREEN",str(int(2))).replace("RED",str(int(1))))
cleanFile.close()
file.close()
    
        
```

```python
from sklearn import linear_model , metrics , tree , cross_validation
from numpy import genfromtxt , savetxt
dataset = genfromtxt(open('trainDataClean_FeatureSelection.csv','r'), delimiter=',', dtype='f8')[1:]
exampleEntry = genfromtxt(open('example_entry.csv','r'),delimiter=',',dtype='f8')[1:]
refIds = [x[0] for x in exampleEntry]
target = [x[1] for x in dataset]
train = [x[2:] for x in dataset]
test = genfromtxt(open('testDataClean.csv','r'), delimiter=',', dtype='f8')[1:]
testData = [x[1:] for x in test]

LogModel = linear_model.LogisticRegression()
LogModel.fit(train,target)
predictionsProb = LogModel.predict_proba(testData)

predictData = [[refIds[index], x[1]] for index, x in enumerate(predictionsProb)]

savetxt('Submissions/submissionLog.csv', predictData, delimiter=',', fmt='%d,%f', 
            header='RefId,IsBadBuy', comments = '')
```
